```python
from random import randrange

def DisplayBoard(board):

    print('+-----------+-----------+----------+')
    print('|           |           |          |')
    print('|     ' +  board[0][0]  + '     |     ' +  board[0][1]  + '     |    ' +  board[0][2]  + '     |')
    print('|           |           |          |')
    print('+-----------+-----------+----------+')
    print('|           |           |          |')
    print('|     ' +  board[1][0]  + '     |     ' +  board[1][1]  + '     |    ' +  board[1][2]  + '     |')
    print('|           |           |          |')
    print('+-----------+-----------+----------+')
    print('|           |           |          |')
    print('|     ' +  board[2][0]  + '     |     ' +  board[2][1]  + '     |    ' +  board[2][2]  + '     |')
    print('|           |           |          |')
    print('+-----------+-----------+----------+')

def EnterMove(board):
    ok = False     
    while not ok:
        move = input("Ingresa tu movimiento: ")
        ok = len(move) == 1 and move >= '1' and move <= '9' 
        if not ok:
            print("Mal movimiento- repite tu entrada") 
            continue
        move = int(move) - 1     
        row = move // 3     
        col = move % 3     
        sign = board[row][col]  
        ok = sign not in ['O','X'] 
        if not ok: 
            print("El campo ya estÃ¡ ocupado - repite tu entrada")
            continue
    board[row][col] = 'O' 

        
def MakeListOfFreeFields(board):
    global espacios_vacios 
    espacios_vacios = [ ]
    
    for fila in range(0,3):
        for columna in range(0,3):
            if board[fila][columna] == 'X' or board[fila][columna] == 'O':
                pass     
            else:
                espacios_vacios.append(([fila],[columna]))
    

def VictoryFor(board, sign):
    
    if sign == 'O':
        print('COMPRUEBA LOS MOVIMIENTOS')
    else:
        print('COMPRUEBA LOS MOVIMIENTOS')
    if board[0][0] == sign and board[0][1] == sign and board[0][2] == sign:
        return True
    if board[1][0] == sign and board[1][1] == sign and board[1][2] == sign: 
        return True
    if board[2][0] == sign and board[2][1] == sign and board[2][2] == sign: 
        return True
    if board[0][0] == sign and board[1][0] == sign and board[2][0] == sign:
        return True
    if board[0][1] == sign and board[1][1] == sign and board[2][1] == sign:
        return True
    if board[0][2] == sign and board[1][2] == sign and board[2][2] == sign:
        return True  
    if board[0][0] == sign and board[1][1] == sign and board[2][2] == sign:
        return True   
    if board[2][0] == sign and board[1][1] == sign and board[0][2] == sign:
        return True   
    

def DrawMove(board):
    while True:
        fila = randrange(3)
        columna = randrange(3)
        
        if ([fila],[columna]) not in espacios_vacios:
            continue
        else:
            board[fila][columna] = 'X'
            return

board = [['1', '2', '3'],['4', 'X', '6'],['7', '8', '9']]
movimientos = 1
usuario = 'O'
maquina = 'X'

print('Bienvenidos a TIC-TAC-TOE.')
print('Buena suerte.')
DisplayBoard(board)
print()
   
while movimientos < 9:
    EnterMove(board)
    movimientos += 1
    DisplayBoard(board)
    
    if VictoryFor(board, usuario) == True:
        print('LE GANASTE A LA COMPUTADORA !!!')
        break
    else:
        print('Espacios vacios')
        MakeListOfFreeFields(board)
        print()
        DisplayBoard(board)
## El codigo se repite    
    print()
    print('Ahora es el turno de la computadora')
    DrawMove(board)
    movimientos += 1
    DisplayBoard(board)
    print()
    
    if VictoryFor(board, maquina) == True:
        print('La computadora te gano, suerte a la proxima...')
        break
    else:
        print('Espacios Vacios son los siguientes')
        MakeListOfFreeFields(board)
        print()
        DisplayBoard(board)
        
else:
    print('Nadie ha ganado')

print('Fin del juego')
```
